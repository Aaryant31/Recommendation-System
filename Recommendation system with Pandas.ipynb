```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
column_names = ['user_id', 'item_id', 'rating', 'timestamp']
df = pd.read_csv('u.data', sep='\t', names=column_names)
```

```python
df.head()
```

```python
movie_titles = pd.read_csv("Movie_Id_Titles")
movie_titles.head()
```

```python
df = pd.merge(df,movie_titles,on='item_id')
df.head()
```

```python
df.groupby('title')['rating'].mean().sort_values(ascending=False).head()
```

```python
df.groupby('title')['rating'].count().sort_values(ascending=False).head()
```

```python
rating_df= pd.DataFrame(df.groupby('title')['rating'].mean())
```

```python
rating_df.head()
```

```python
rating_df['num of rating']=df.groupby('title')['rating'].count()
```

```python
rating_df.head()
```

```python
rating_df.sort_values('num of rating',ascending=False).head()
```

```python
rating_df['num of rating'].hist(bins=80)
```

```python
rating_df['rating'].hist(bins=70)
```

```python
sns.jointplot(x='rating',y='num of rating',data=rating_df,alpha=0.5)
```

```python
movie_metrix=df.pivot_table(index='user_id',columns='title',values='rating')
```

```python
movie_metrix.head()
```

```python
rating_df.sort_values('num of rating',ascending=False).head(10)
```

```python
starwars_user_ratings = movie_metrix['Star Wars (1977)']
liarliar_user_ratings = movie_metrix['Liar Liar (1997)']
starwars_user_ratings.head()
```

```python
similar_to_starwars = movie_metrix.corrwith(starwars_user_ratings)
similar_to_liarliar = movie_metrix.corrwith(liarliar_user_ratings)
```

```python
similar_to_starwars
```

```python
corr_starwars = pd.DataFrame(similar_to_starwars,columns=['Correlation'])
corr_starwars.dropna(inplace=True)
corr_starwars.head()
```

```python
corr_starwars.sort_values('Correlation',ascending=False).head(10)
```

```python
corr_starwars = corr_starwars.join(rating_df['num of rating'])
corr_starwars.head()
```

```python
corr_starwars[corr_starwars['num of rating']>100].sort_values('Correlation',ascending=False).head()
```

```python
corr_liarliar = pd.DataFrame(similar_to_liarliar,columns=['Correlation'])
corr_liarliar.dropna(inplace=True)
corr_liarliar = corr_liarliar.join(rating_df['num of rating'])
corr_liarliar[corr_liarliar['num of rating']>100].sort_values('Correlation',ascending=False).head()
```
